```python
# Fetch the connectivity table
_, df_connections = fetch_adjacencies() 
```

```python
# Fetch the neuron table with annotations
bodyIds = list(dict.fromkeys(df_connections.bodyId_pre.tolist() + df_connections.bodyId_pre.tolist()))
criteria = NC(bodyId=bodyIds)
df_classifications, _ = fetch_neurons(criteria) 
```

```python
# Trim dataframes to traced neurons
bodyIds_traced = df_classifications.bodyId[df_classifications.status=='Traced'].tolist()
df_connections = df_connections[(df_connections.bodyId_pre.isin(bodyIds_traced)) & (df_connections.bodyId_post.isin(bodyIds_traced))]
df_classifications = df_classifications[df_classifications.bodyId.isin(bodyIds_traced)]
```

```python
# Trim classification dataframe to relevant information
columns_to_keep = ['bodyId','instance','class','subclass','predictedNt','predictedNtProb','hemilineage','synonyms','tag','description']
df_classifications = df_classifications.loc[:,df_classifications.columns.intersection(columns_to_keep)]
```

```python
# Save dataframes as csv
df_connections.to_csv('manc_v121_connections.csv',index=False)
df_classifications.to_csv('manc_v121_classifications.csv',index=False)
```
