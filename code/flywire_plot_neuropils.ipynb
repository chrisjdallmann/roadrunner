```python
# Import libraries 
import navis
import flybrains
from fafbseg import flywire
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Settings
neuropils = "RRN"
```

```python
# Get brain mesh 
brain = flybrains.FLYWIRE

# Get neuropil meshes
if neuropils == "RRN":
    neuropils = flywire.get_neuropil_volumes(['ICL_L', 'ICL_R', 'IB_L', 'IB_R']) 
else:
    neuropils = flywire.get_neuropil_volumes(['ICL_L', 'ICL_R', 'PLP_L', 'PLP_R', 'SMP_L', 'SMP_R']) 
```

```python
# Set colors
color = []
color.append([0.75,0.75,0.75,0.2]) # Brain

# Neuropils 
for neuropil in neuropils:
    if neuropils == "RRN":
        color.append([0/255,173/255,238/255,0.5])
    else:
        color.append([36/255,62/255,144/255,0.5]) 
```

```python
# Plot neuropils 
fig, ax = navis.plot2d(
    [brain.mesh, neuropils], 
    dps_scale_vec=1000,    
    method='2d', 
    view=('-x','-y'), 
    figsize=(12, 12),
    color=color, 
    )

#plt.savefig(r'C:/Users/Chris/Desktop/figure.png', dpi=400)
plt.show()
```
