```python
# Import libraries 
import navis
import flybrains
from fafbseg import flywire
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get brain mesh 
brain = flybrains.FLYWIRE

# Get neuropil meshes
neuropils = flywire.get_neuropil_volumes(['FLA_R', 'VES_R']) 
```

```python
# Set colors
color = []

# Brain
color.append([0.75,0.75,0.75,0.2]) 

# Neuropils 
for neuropil in neuropils:
    color.append([0/255,173/255,238/255,0.5])
```

```python
# Plot neuropils 
fig, ax = navis.plot2d(
    [brain.mesh, neuropils], 
    dps_scale_vec=1000,    
    method='2d', 
    view=('-x','-y'), 
    figsize=(12, 12),
    color=color, 
    )

plt.show()
```
